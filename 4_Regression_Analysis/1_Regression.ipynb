# Regression Analysis(회귀 분석)

* In statistical modeling, regression analysis is a set of statistical processes for **estimating the relationships between a dependent variable** (often called the 'outcome' or 'response' variable) **and one or more independent variables** (often called 'predictors', 'covariates', 'explanatory variables' or 'features').

from [wikipedia](https://en.wikipedia.org/wiki/Regression_analysis)

* **Goal: Find most suitable linear/nonlinear equation for observed values.**

## History
* Francis Galton's 1875 illustration of the correlation between the heights of adults and their parents. The observation that adult children's heights tended to deviate less from the mean height than their parents suggested the concept of   "regression toward the mean", giving regression its name.

# Linear Regression

![](regression_analysis_sample.png)

Linear Equation derived from linear regression analysis might be in following form : $$y = \alpha + \beta x$$

Then, $y_i$, the arbitrary element of set $Y$ can be expressed as: $$y_i = \alpha + \beta x_i + \epsilon_i$$

$\epsilon_i$ is error (the difference between $y$ (predicted value from equation) and $y_i$(observed value))

Thus, the set of $y$, $Y$ can be expressed as: $$Y =  \alpha + \beta x + \epsilon$$

Then, the goal will be **Find the equation that has least error $\epsilon$**, which can be rewritten as either 'minimize $\sum \lvert \epsilon_i \rvert$' or **'minimize $\sum {\epsilon_i}^2$'**

($\sum {\epsilon_i}^2$ is preferred because it is easier mathmathically)

# Least Square Method: 최소제곱법 (for simple regression analysis, 단일변수 회귀분석을 위한 방법)
## Find equation that has minimized SSE(Sums of Squared Error: $\sum {\epsilon_i}^2$)
* SSE is cost function
$$SSE=\sum (\epsilon_i)^2=\Sigma(Y_i-\alpha-\beta x_i)^2$$

Since SSE is expressed as quadrantic equation, mimimum value might be local minimum.
Thus, Least Square Estimation for $\alpha$,$\beta$ would be:

<br>

$$\frac{\partial SSE}{\partial \alpha} = 2 \sum(Y_i - \alpha - \beta x_i) (-1) = 0 \quad \cdots \quad (1)$$

<br>

$$\frac{\partial SSE}{\partial \beta} = 2 \sum(Y_i - \alpha - \beta x_i) (-x_i) = 0  \quad \cdots \quad (2)$$

$(1)$ can be rewritten as:

<br>

$$\sum(Y_i - \alpha - \beta x_i)=0$$

<br>

$$\sum Y_i = n \alpha - \beta \sum x_i \quad \cdots \quad (3)$$

<br>

$$\overline{x} \sum Y_i = \alpha \sum x_i - \beta \overline{x} \sum x_i \quad \cdots \quad (3)\times\overline{x}$$

<br>

$$\overline{Y} = \alpha + \beta \overline{x} \quad \cdots \quad (3) \div n$$

<br>

$$(\because \overline{x} = \frac{1}{n} \sum_{i=1}^{n} x_i)$$ 

<br>

also, $(2)$ can be rewritten as:

<br>

$$\sum(Y_i - \alpha - \beta x_i)(x_i)=0$$

<br>

$$\sum x_i Y_i = \alpha \sum x_i + \beta \sum {x_i}^2 \quad \cdots \quad (4)$$

<br>

Then, $(4)-(3)\times\overline{x}$ derives following:

<br>

$$\sum (x_i Y_i - n\overline{x}\overline{Y}) = \beta (\sum {x_i}^2 - n {\overline{x}}^2)$$

<br>

$$\beta = \frac{\sum (x_i Y_i - n\overline{x}\overline{Y})}{\sum {x_i}^2 - n {\overline{x}}^2} = \frac{\sum (x_i-\overline{x})(Y_i-\overline{Y})}{\sum({x_i-\overline{x}})^2}$$

<br>

$$(\because \sum (x_i-\overline{x})(Y_i-\overline{Y})=\sum (x_i Y_i - n\overline{x}\overline{Y})), \sum {x_i}^2 - n {\overline{x}}^2 = \sum({x_i-\overline{x}})^2)$$ 

<br>

Thus, $\alpha$ and $\beta$ for minimized SSE would be

<br>

$$\therefore \quad \hat{\beta} = \frac{\sum (x_i-\overline{x})(Y_i-\overline{Y})}{\sum({x_i-\overline{x}})^2}, \quad \hat{\alpha}=\overline{Y} - \hat{\beta} \overline{x}$$

# Multiple Regression Analysis

$$Y_i = \beta_0 + \beta_1 x_{1i} + \beta_2 x_{2i} + \cdots + \beta_k x_{ki} + \epsilon_i, \quad \epsilon_i \sim N(0,{\sigma}^2), \quad i=1,\cdots,n$$

<br>

$$
X = 
\begin{pmatrix}
1 & x_{11} & x_{21} & \cdots & x_{k1} \\
1 & x_{12} & x_{22} & \cdots & x_{k2} \\
\vdots & \vdots & \vdots  & \ddots & \vdots  \\
1 & x_{1n} & x_{2n} & \cdots & x_{kn} 
\end{pmatrix} ,\quad
\beta = \begin{pmatrix} \beta_0 \\ \beta_1 \\ \vdots \\ \beta_k \end{pmatrix}, \quad
Y= \begin{pmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{pmatrix} = X \beta + \epsilon, \quad
$$

<br>

Find $\beta$ for minimized SSE:

<br>

$$SSE = \sum (\epsilon_i)^2 = \epsilon^T \epsilon = (Y-X\beta)^T(Y-X\beta) = (Y^T-\beta^TX^T)(Y-X\beta)$$

<br>

$$=Y^T Y - Y^T X\beta - \beta^T X^T Y + \beta^T X^T X\beta$$

<br>

$$=Y^T Y - 2 \beta^T X^T Y + \beta^T X^T X\beta \quad (\because (AB)^T = B^T A^T)$$

<br>

$$\frac{\partial SSE}{\partial \beta} = -2(X^T Y) -2(X^T X)\hat{\beta}=0$$

<br>

$$\therefore \quad \hat{\beta} = (X^T X)^{-1} X^T Y$$

````python
# Assignment 1
# for X = np.array([1,2,3]), Y = np.array([1,3,2])
# find the linear regression equation and plot scatter and line. (use simple linear regression)

import numpy as np

X = np.array([1,2,3])
Y = np.array([1,3,2])
n = len(X)

Xbar = np.mean(X)
Ybar = np.mean(Y)

numerator = denominator = 0
for i in range(n):
    numerator += (X[i] - Xbar) * (Y[i] - Ybar)
    denominator += (X[i] - Xbar)**2

beta = numerator / denominator
alpha = Ybar - beta*Xbar

print("Alpha: %g / Beta: %g"%(alpha, beta))

import matplotlib.pyplot as plt
domain=np.linspace(0,4,10)
plt.scatter(X,Y)
plt.plot(domain,alpha+domain*beta)
plt.show()
````

````python
# Assignment 2
# for X = np.array([1,2,3]), Y = np.array([1,3,2])
# find the linear regression equation. (use multiple linear regression)
import numpy as np

X = np.array([[1,1], [1,2], [1,3]])
Y = np.array([1,3,2])

tmp1 = np.linalg.inv(np.dot(X.T,X)) 
tmp2 = np.dot(X.T,Y)

cof = np.dot(tmp1,tmp2)
print("Alpha: %g / Beta: %g" %(cof[0],cof[1]))
````

# Logistic Regression
* is a statistical model that in its basic form uses a logistic function to model a binary dependent variable, although many more complex extensions exist. 
* is estimating the parameters of a logistic model which has a dependent variable with two possible values, such as pass/fail

from [wikipedia](https://en.wikipedia.org/wiki/Logistic_regression)

* General type of regression is not suitable for binary model
* Generally regression $\hat{y}$'s range is $(-{\infty},{\infty})$, while desired range is only $\{0,1\}$

## Implement Logistic Function

$$Y=\frac{1}{1+e^{-x}}$$

<br>

Sigmoid Function (one of Logistic Functions) $Y=\frac{1}{1+e^{-x}}$ satisfies $0 < Y <1$

<br>

$$\widehat{Y}=\frac{1}{1+exp(-X \cdot \hat{\beta})}$$

<br>

$$\frac{\widehat{Y}}{1-\widehat{Y}}=\frac{1}{exp(-X \cdot \hat{\beta})} = exp(X \cdot \hat{\beta})$$

<br>

$$ln(\frac{p}{1-p})= X \cdot \beta = \beta_0 + \beta_1 x_1 + \cdots + \beta_k x_k \qquad p=Pr(Y=1)$$

<br>

$$ln(\frac{p_i}{1-p_i})= X \cdot \beta \qquad p_i=Pr(Y=1)$$

<br>

For given $\hat{\beta}$, $\widehat{Y}$ can be estimated as follows: 

<br>

$$\widehat{Y}=\frac{1}{1+exp(-X \cdot \hat{\beta})} = \frac{1}{1+exp(-X \cdot \hat{\beta})} \cdot \frac{exp(X \cdot \hat{\beta})}{exp(X \cdot \hat{\beta})} = \frac{exp(X \cdot \hat{\beta})}{1+exp(X \cdot \hat{\beta})}$$

## Cost minimizing (using Cost function (Loss Function))

* Cost function decreases if it is correct, while increases if it is incorrect.
* Cost function satisfies convexity(볼록성: 아래로 볼록한 함수)

For Bias $B$ and Weight $W$ ($\beta$ in statistics), following Logistic function (Sigmoid) $H(X)$ and Cross Entropy function, Cost Function $Cost(W)$ is given.


$$H(X)=\frac{1}{1+exp(-WX)}$$

<br>

$$
Cost(W)=
    \begin{cases} 
    \sum -ln(H(X)) & \,\,\,\,\,\,\,\,\,(y=1)\\
    \sum -ln(1-H(X)) & \,\,\,\,\,\,\,\,\, (y=0)
    \end{cases}
$$


Cost Function can be rewritten in one line:

$$Cost(W)=\sum \big[ -y_iln(H(X)) + (1-y_i) -ln(1-H(X)) \big]$$

````python
# Convexity Check
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.array([1,2,3,4,5,6])
y = np.array([0,0,0,1,1,1])
xy = np.array([[1,0],[2,0],[3,0],[4,1],[5,1],[6,1]])

n = 100
a = np.linspace(-5,5,n)
b = np.linspace(-2.5,2.5,n)
a,b = np.meshgrid(a,b)
cost1 = np.zeros((n,n)) # cross entropy function

for val in xy:
    temp = -val[1]*np.log(sig(a+b*val[0])) - (1-val[1])*np.log(1-sig(a+b*val[0]))
    cost1 += temp

cost2 = np.zeros((n,n))
for val in xy:
    temp = (val[1]-sig(a+b*val[0]))**2 # error sums of square function
    cost2 += temp

ax = plt.axes(projection='3d')
ax.contour3D(a,b,cost1,n,cmap='binary') # satisfies convexity
#ax.contour3D(a,b,cost2,n,cmap='binary') # SSE failed convexity
````

Cross Entropy Function is known as **Maximum Likelihood Function for Logistic Regression**. (Maximum Likelihood Function: 최대 가능도 함수; 가장 그럴싸 하다)


## Derivation of MLF for Multiple Logistic Regression

Cross Entropy 함수는 Logistic Regression의 Maximum Likelihood 함수다.
최대 가능도 함수(Maximum Likelihood function)에 대해 알아보자.

$Y_i = \frac{1}{ 1+exp( - {\beta}_0 - {\beta}_1 x_i + \epsilon ) }$의 식에서 $Y_i$는 0 혹은 1의 값을 갖는 확률변수다.

그러므로 $Y_i \sim Bernoulli(p_i)$이다. 여기서, $p_i=Pr(Y_i = 1)$ 이고 확률질량함수(P.M.F.)는 아래와 같다.

$$f(y)= P(Y=y) = {p_i}^y (1 - p_i  )^{1-y}$$


만약, $Y_i|X_i = [y_1 , y_2 , \cdots , y_n ]$ 이 관측되었다면 이렇게 관측될 확률은 아래와 같다.


$$\prod_{i=1}^{n} {p_i}^{y_i}(1-p_i)^{1-y_i}$$


이 수식을 미지 모수 $p_i$에 대한 최대 가능도 함수라고 한다.
위 수식에서 $y_i$ 는 주어진 값이고 미지수는 오직 $p_i$ 뿐이다.

이 함수를 최대화 하는 $p_i$ 가 가장 관측값을 잘 설명하는 값이라고 생각하는 것이 MLE 추정(Maximum Likelihood
Estimation)이다.

$p_i$ 는 $y_i$ 에 대한 추정값으로 $p_i>0.5$ $\rightarrow$ $y_i =1$ 아니면 $y_i=0$이 되므로 아래와 같은 식이 성립한다.


$$p_i=\frac{1}{1+exp(-\hat{\beta_0}-\hat{\beta_1}x_i)}, \quad L(p_i)=\prod_{i=1}^{n} {p_i}^{y_i}(1-p_i)^{1-y_i}$$


이 식을 최대화 하는 $\beta_0$, $\beta_1$이 우리가 구하고자 하는 미지수이고 이를 MLE(Maximum Likelihood Estimator)라고
한다. Likelihood 함수에 로그를 취하면, 아래와 같다.


$$ln L(\beta_0,\beta_1) = \sum_{i=1}^{n} \Big[ y_i ln\frac{1}{1+exp(-\hat{\beta_0}-\hat{\beta_1}x_i)} + (1-y_i)ln(1-\frac{1}{1+exp(-\hat{\beta_0}-\hat{\beta_1}x_i)}) \Big]$$

<br>

$$= \sum_{i=1}^{n} \Big[ y_i lnp_i + (1-y_i)ln(1-p_i) \Big]$$


이 식에 “-1”을 곱한 식이 Cross Entropy 함수다.
최대가능도 함수를 최대화하는 것이나 Cross Entropy 함수를 최소화하는 것이나 같은 것이다.
이 함수를 최소화하는 $\beta_0$, $\beta_1$ 를 구하면 MLE가 된다.

## Gradient Descent Algorithm:  경사하강법

$$w:=w-\eta \nabla Q_{i}(w)$$

which can be rewritten to:

$$B := B - \eta \frac{\partial}{\partial B} Cost(W,B)$$

<br>

$$W := W - \eta \frac{\partial}{\partial W} Cost(W,B)$$

$B$: Bias<br>
$W$: Weight<br>
$Q(W)$: Cost Function<br>
$Cost(W,B)$: Cost Function<br>
$\eta$: learning rate


# Assignment 3
* For following Cost Function (Cross Entropy Function), derive partial derivative function for bias $b$ and weight $w$
$ Cost(w,b) = -\sum \Big[ y_i ln(S(wx_i+b)) +(1-y_i)ln(1-S(wx_i+b)) \Big]$

$pf)$

$$S(wx_i+b) = \frac{1}{1+exp(-wx_i-b)}$$

<br>

$$\frac{\partial}{\partial b} S(wx_i+b) = \frac{exp(-wx_i-b)}{1+exp(-wx_i-b)} = S(wx_i+b)(1-S(wx_i+b))$$

<br>

$$\frac{\partial}{\partial w} S(wx_i+b) = \frac{exp(-wx_i-b)x_i}{1+exp(-wx_i-b)} = S(wx_i+b)(1-S(wx_i+b))x_i$$

<br>

$Thus,$

<br>

$$ \frac {\partial }{\partial b} y_i ln(S(wx_i+b)) = y_i(1-S(wx_i+b))$$

<br>

$$ \frac {\partial }{\partial b} (1-y_i)ln(1-S(wx_i+b)) =-(1- y_i)S(wx_i+b))$$

<br>

$$ \frac {\partial }{\partial w} y_i ln(S(wx_i+b)) = y_i(1-S(wx_i+b))x_i$$

<br>

$$ \frac {\partial }{\partial w} (1-y_i)ln(1-S(wx_i+b)) =-(1- y_i)S(wx_i+b))x_i$$

<br>

$In \  conclusion,$

<br>

$$ \frac {\partial }{\partial b}Cost(w,b) =  -\sum \Big[y_i - S(wx_i+b)\Big] $$

<br>

$$ \frac {\partial }{\partial w}Cost(w,b) =  -\sum \Big[y_i - S(wx_i+b)\Big] x_i$$

````python
# Assignment 4

# For x = np.array([1,2,3,4,5,6]) and y = np.array([0,0,0,1,1,1]) 
# Find bias b and weight w using Gradient Descent Algorithm

def sigmoid(x):
    return 1/(1+np.exp(-x))

def deriv(bw): # derivatives: 도함수 / bw=[bias, weight]
    partial_b = (-1) * np.sum(y - sigmoid(bw[0] + bw[1]*x)) 
    partial_w = (-1) * np.sum((y - sigmoid(bw[0] + bw[1]*x))*x)
    return np.array([partial_b, partial_w])

x = np.array([1,2,3,4,5,6])
y = np.array([0,0,0,1,1,1])
bw = np.array([0.5,0.5]) # random initial value

np.set_printoptions(suppress=True, precision=3)
alpha = 0.1 # learning rate
for i in range(2001):
    bw = bw - alpha * deriv(bw)
    if i % 100 == 0: 
        print("%d of 2000 | Bias: %g, Weight: %g" %(i,cof[0],cof[1]))
        yhat = sigmoid(bw[0] + bw[1]*x)
        print(yhat)
````
